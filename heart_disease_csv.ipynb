```python
# feature and target separation
# feature
x=df.drop('target',axis=1)
# target
y=df['target']
```

```python
x.columns
```

```python
y.name
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
y.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
x.shape
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
# feature scaling - preprocessing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
# here we will use logistic regression cuz it is a binary
# classification problem
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_train = lr.predict(x_train)
y_pred_test = lr.predict(x_test)

print('train acc:',accuracy_score(y_train,y_pred_train))
print('test acc:',accuracy_score(y_test,y_pred_test))
```

```python
# predictions
y_pred_train=lr.predict(x_train)
y_pred_test=lr.predict(x_test)
```

```python
# accuracy check
from sklearn.metrics import accuracy_score
print('train acc:',accuracy_score(y_train,y_pred_train))
print('test acc:',accuracy_score(y_test,y_pred_test))
```

```python
y_train[:5]
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_test)
cm
```

```python
# classification report
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred_test)
print(cr)
```

```python
# next model we using here is SVM ( supports vector machines)
# it finds a strong,margin between two classes
# it finds a strong margin based decesion boundry that often
# improves recall on small well scalled dataset
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
# prediction
y_pred_train=svc.predict(x_train)
y_pred_test=svc.predict(x_test)
```

```python
# evaluating
svc.score(x_train,y_train)
```
