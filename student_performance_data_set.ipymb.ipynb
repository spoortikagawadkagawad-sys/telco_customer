```python
# separate numerical and categorical column
numerical_columns=df.select_dtypes(include=[np.number])
categorical_columns=df.select_dtypes(exclude=[np.number])
```

```python
print('numerical columns:',numerical_columns.columns)
print('categorical columns:',categorical_columns.columns)
```

```python
# tarin and spilt test
from sklearn.model_selection import train_test_split
x = numerical_columns
y = df['g3']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
df.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# first ml model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# predictions
y_pred = model.predict(x_test)
```

```python
# evalution matrices
# MAE Mean = mean absolute error
# error which is the difference between actual price
# from the predicted price
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,y_pred)
print('MAE:',mae)
```

```python
# R2 score
# it defines how well the model fits the data
# <0.5 not fitting well the data
# 0.5 or near - avg value for fit
# >0.7 - fits well the data
# 1.0 - model is perfect
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print('R2:',r2)
```
